# Machine Learning Algorithm 

#### Here In Our Data The Best Model is LogisticRegrssion due to it's more better for non-numeric ,
#### Also ABout The Accuracy Logistic Regression Will Be High Accuracy in our data 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_final, df['stars'], test_size=0.2, random_state=42)
```

```python
# Train
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Extra Work..How To Increase Our Model Accuracy

```python
# Binary classification target: 1 if stars >=4, else 0
df['binary_stars'] = df['stars'].apply(lambda x: 1 if x >= 4 else 0)

# Text vectorization using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_text = tfidf_vectorizer.fit_transform(df['text'])
X_text_df = pd.DataFrame(X_text.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Scale numeric features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_features = df[['cool', 'funny', 'useful']]
num_scaled = scaler.fit_transform(num_features)
num_scaled_df = pd.DataFrame(num_scaled, columns=['cool', 'funny', 'useful'])

# Combine text and numeric features
X_final = pd.concat([X_text_df, num_scaled_df], axis=1)

# Target variable
y = df['binary_stars']
```

```python
# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
# Train Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Predict and evaluate
from sklearn.metrics import classification_report, accuracy_score
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Perfect Accuracy due to Classifying it it was the right choice for understanding data and predict in it
